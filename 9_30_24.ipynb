```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Data
data = {
    'Age': [3, 4, 5, 6, 7, 8, 9, 11, 12, 14, 15, 16, 17],
    'Wing Length': [1.4, 1.5, 2.2, 2.4, 3.1, 3.2, 3.2, 3.9, 4.1, 4.7, 4.5, 5.2, 5.0]
}

# Create DataFrame
df = pd.DataFrame(data)

# Fit the regression model
X = sm.add_constant(df['Age'])  # Add a constant for the intercept
model = sm.OLS(df['Wing Length'], X).fit()
df['Predicted Wing Length'] = model.predict(X)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Wing Length'], color='blue', label='Data points')
plt.plot(df['Age'], df['Predicted Wing Length'], color='red', label='Regression Line')
plt.title('Relationship between Age and Wing Length with Regression Line')
plt.xlabel('Age')
plt.ylabel('Wing Length')
plt.grid()
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Data
data = {
    'Age': [3, 4, 5, 6, 7, 8, 9, 11, 12, 14, 15, 16, 17],
    'Wing Length': [1.4, 1.5, 2.2, 2.4, 3.1, 3.2, 3.2, 3.9, 4.1, 4.7, 4.5, 5.2, 5.0]
}

# Create DataFrame
df = pd.DataFrame(data)

# Fit the regression model
X = sm.add_constant(df['Age'])  # Add a constant for the intercept
model = sm.OLS(df['Wing Length'], X).fit()

# Get the predicted values and the confidence intervals
predictions = model.get_prediction(X)
pred_summary = predictions.summary_frame(alpha=0.05)  # 95% CI

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Wing Length'], color='blue', label='Data points')
plt.plot(df['Age'], pred_summary['mean'], color='red', label='Regression Line')
plt.fill_between(df['Age'], pred_summary['obs_ci_lower'], pred_summary['obs_ci_upper'], color='pink', alpha=0.3, label='95% CI')
plt.title('Regression Line with 95% Confidence Intervals')
plt.xlabel('Age')
plt.ylabel('Wing Length')
plt.grid()
plt.legend()
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm

# Data
data = {
    'Age': [3, 4, 5, 6, 7, 8, 9, 11, 12, 14, 15, 16, 17],
    'Wing Length': [1.4, 1.5, 2.2, 2.4, 3.1, 3.2, 3.2, 3.9, 4.1, 4.7, 4.5, 5.2, 5.0]
}

# Create DataFrame
df = pd.DataFrame(data)

# Fit the regression model
X = sm.add_constant(df['Age'])  # Add a constant for the intercept
model = sm.OLS(df['Wing Length'], X).fit()

# Get R-squared value
r_squared = model.rsquared

r_squared
```

	•	The  R^2  value will range from 0 to 1:
	•	0 indicates that the model does not explain any variability in the dependent variable.
	•	1 indicates that the model explains all the variability in the dependent variable.
	•  A higher  R^2  value suggests that the model explains a greater proportion of the variance in the dependent variable.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Original Data
data = {
    'Age': [3, 4, 5, 6, 7, 8, 9, 11, 12, 14, 15, 16, 17],
    'Wing Length': [1.4, 1.5, 2.2, 2.4, 3.1, 3.2, 3.2, 3.9, 4.1, 4.7, 4.5, 5.2, 5.0]
}

# Create DataFrame
df = pd.DataFrame(data)

# Fit the regression model on the original data
X_original = sm.add_constant(df['Age'])  # Add a constant for the intercept
model_original = sm.OLS(df['Wing Length'], X_original).fit()

# Predictions for the original data
predictions_original = model_original.predict(X_original)

# Add noise to the Wing Length
np.random.seed(42)  # For reproducibility
noise = np.random.normal(0, 0.2, size=df['Wing Length'].shape)  # Add Gaussian noise
df['Noisy Wing Length'] = df['Wing Length'] + noise

# Fit the regression model on the noisy data
X_noisy = sm.add_constant(df['Age'])  # Add a constant for the intercept
model_noisy = sm.OLS(df['Noisy Wing Length'], X_noisy).fit()

# Predictions for the noisy data
predictions_noisy = model_noisy.predict(X_noisy)

# Plotting
plt.figure(figsize=(12, 8))
plt.scatter(df['Age'], df['Wing Length'], color='blue', label='Original Data', alpha=0.6)
plt.scatter(df['Age'], df['Noisy Wing Length'], color='orange', label='Noisy Data', alpha=0.6)
plt.plot(df['Age'], predictions_original, color='green', label='Regression Line (Original Data)')
plt.plot(df['Age'], predictions_noisy, color='red', label='Regression Line (Noisy Data)')
plt.title('Effect of Noise on Linear Regression')
plt.xlabel('Age')
plt.ylabel('Wing Length')
plt.grid()
plt.legend()
plt.show()

# Print the summaries of the regression models
print("Original Data Regression Summary:")
print(model_original.summary())
print("\nNoisy Data Regression Summary:")
print(model_noisy.summary())
```
